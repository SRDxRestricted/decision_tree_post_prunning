```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris
```

```python
iris.target
```

```python
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
```

```python
df.head()
```

```python
X = df.iloc[: , :-1]
y = iris.target
```

```python
y
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state = 42)
```

```python
X_train.head()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
#by making a decison tree by default it takes all the hyperparameters as none
treemodel = DecisionTreeClassifier()
```

```python
treemodel.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treemodel, filled = True)
```

```python
#PostPrunning
treemodel_pruned = DecisionTreeClassifier(max_depth = 2)
treemodel_pruned.fit(X_train, y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (15,10))
tree.plot_tree(treemodel_pruned, filled = True)
```

```python
y_pred = treemodel_pruned.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
score = accuracy_score(y_test, y_pred)
classification = classification_report(y_test, y_pred)
```

```python
print(score)
```

```python
print(classification)
```
